```python
import pandas as pd
import numpy as np
from nsepy import get_history
from datetime import date
```

```python
# Buy if the gap between day's high & low is atleast 1% and gap between day's close & high is atleast 0.3%.
# Square off on next day if gap between close & high is atleast 1% or square off at closing
```

```python
df = get_history(symbol="TECHM",start=date(2019,11,20),end=date(2021,11,20))
```

```python
# remove special character
df.columns = df.columns.str.replace('[%]', 'Perc')
```

```python
df
```

```python
df['Buy'] = np.where((df.High > (1.01 * df.Low)) &
                     (df.High > (1.007 * df.Close)),1,0)
```

```python
df
```

```python
df['Buyprice'] = np.where(df.Buy == 1, df.Close, 0)
```

```python
df['Sellprice'] = np.where(df.High.shift(-1)>1.01*df.Buyprice,1.01*df.Buyprice,df.Close.shift(-1))
```

```python
df.tail(20)
```

```python
x = df[(df.Buy == 1)]
```

```python
x
```

```python
profits = ((x.Sellprice - x.Buyprice)/x.Buyprice)
```

```python
profits.tail(30)
```

```python
profits = profits[:-1]
```

```python
def backtest(arr):
    winrate = len(arr[arr > 0])/len(arr)
    max_dd = min(arr)
    mean = arr.mean()
    median = arr.median()
    gain = (arr + 1).cumprod()[-1]
    return winrate, max_dd, mean, median, gain
```

```python
backtest(profits)
```
